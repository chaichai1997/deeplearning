```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('test.jpg')
```

```python
# 访问像素值
px = img[100, 100]
print(px)
```

```python
# 修改像素值
img[100, 100] = 255, 255, 255
print (img[100, 100])
```

```python
# 访问图像属性，返回图像的行列通道数
print(img.shape)
```

```python
# 返回图像的像素点个数
print(img.size)
# 返回图片数据类型
print(img.dtype)
```

```python
# image ROI region of images 区域图像
# ball = img[200:300, 200:300]
# img[300:400, 300:400] = ball
```

```python
# Splitting and Merging Image Channels分割和合并图像通道
# 拆分
b, g, r = cv2.split(img)
plt.subplot(311)
plt.imshow(r)
plt.subplot(312)
plt.imshow(g)
plt.subplot(313)
plt.imshow(b)
```

```python
# 合并
img1 = cv2.merge((b, g, r))
plt.imshow(img1)
```

```python
# image addition 图像加法
x = np.uint8([250])
y = np.uint8([10])
# 250 + 10 => 255
print(cv2.add(x, y))
# 250 + 10=>260 %256 =4
print(x+y) 
```
